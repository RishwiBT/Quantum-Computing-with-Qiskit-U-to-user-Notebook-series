To focus our learning even further, let's examine the hardware itself. Computers store binary information using **transistors**, which essentially showcase the presence of, or absence of, electrons. If electrons are present, the transistor registers the binary bit as `1`. If the electrons are not present, the transistor registers the binary bit as `0`.

To make computers more powerful, we need to add more binary bits. To add more binary bits, we need to add more transistors. To add more transistors, we need more space. We can only make transistors *so* small. The most advanced supercomputer, developed by IBM, has 2 nm transistor technology, with anywhere between 5 to 20 atoms per transistor.

At a certain point, we hit a limit to how small we can make transistors. This, in effect, puts a limit on how powerful our computers can get.

**Could you use a single atom, or electron, as a bit of information?**

Now *this* takes us into the **quantum realm**.

Using pieces of information that are the size of a single atom means we enter a different world of possibilities. At this size, **quantum mechanics** take over. Under the laws of quantum mechanics, particles behave weirdly, possessing not-so-intuitive characteristics. Remember that electrons or photons can act as our data at this quantum scale. A **qu**antum **bit** of data (AKA, a **qubit**) can behave differently than the binary bits of a classical transistor.

* **For example:** A qubit can be put into "superposition", which gives it the ability to have a collection of possible answers, like a shaken-up magic 8 ball. We won't know exactly which answer it displays until we look at it.

That's not all. Using other properties like entanglement and amplification - quantum computers present a totally new paradigm for how to approach and solve problems. ## The Math of Quantum Computation

To study anything, we need a common language to communicate our ideas and concepts about a topic. Mathematics has always been a language that helps us understand physics in depth. Let's start by introducing the basic mathematical notations of quantum information.

A qubit is mathematically denoted using the following StateVector format :

$
\alpha |0 \rangle + \beta |1 \rangle \ \ \rightarrow
\begin{bmatrix}
\alpha \\
\beta
\end{bmatrix}
$

This mathematical expression tells you that upon measurement, the qubit will be found in state $|0\rangle$ with $|\alpha|^2$ probability and in state $|1\rangle$ with $|\beta|^2$ probability. $\alpha$ and $\beta$ are called probability amplitudes.

if $\alpha = \beta = 1/ \sqrt(2) \ \ $ then
$
\begin{bmatrix}
\frac{1}{\sqrt2} \\
\frac{1}{\sqrt2}
\end{bmatrix}
\rightarrow \frac{1}{\sqrt2}
\begin{bmatrix}
1 \\
0
\end{bmatrix}
 \ + \
\frac{1}{\sqrt2}
\begin{bmatrix}
0 \\
1
\end{bmatrix}\ \  \rightarrow \frac{1}{\sqrt2} |0\rangle + \frac{1}{\sqrt2} |1\rangle \ \ $

And now $\alpha^2 = (\frac{1}{\sqrt2})^2 \ and \ \beta^2 = (\frac{1}{\sqrt2})^2$ which means the qubit has a probability of {1/2, 1/2} (i.e 50% and 50%) of being found in either $|0\rangle$ or $|1\rangle$.

This is called quantum superposition and is one of the core phenomena that powers quantum computers.

The obvious question that many might have at this point is, *Why do quantum particles behave that way?*

Let's try to visualize the state vectors using code.

`StateVector()` is the function, in `qiskit.quantum_info`, that you use to input your values to form a vector.

 The `draw()` function is a property of a `StateVector()` that can display the vector in latex or text format.

```python
u = Statevector([1/sqrt(2), 1/sqrt(2)])
display(u.draw("latex"))                      #Displaying the State Vector in latex equation format
display(u.draw("text"))                       #Displaying it in vector format
```

Even though the state of "u"  is in equal superposition of $|0\rangle$ and $|1\rangle$, it will RANDOMLY collapse into one of those states when measured.

Use the `measure()` function to measure the statevector u.

```python
u.measure()
```

<h2>Every time you run, it collapses to different state</h2>

That means, once you measure a quantum state the superposition is lost and the state drops down to one of the states. So to measure it again, the same state should be initiated and measured.

Let's do it 1000 times and plot the histogram with number of times the state collapsed to 0 and 1.

```python
from qiskit.visualization import *

counts = u.sample_counts(1000)            # sample_counts method initializes and measures the state, c times. Here c= 1000
display(counts)
plot_histogram(counts)                    # Plot_histogram function is in qiskit.visualization
```

The Vertical axis is the Z axis. x and y axis are as per marked.

$|0\rangle$ and $|1\rangle$ are the eigenstates of Z so they are on the either sides of the Z axis.

As all the qubits are initialized to $|0\rangle$ by default. The state vector(pink arrow in the above Bloch sphere) points towards the north pole($|0\rangle$)

A Qubit can be in superposition(linear combination) of $|0\rangle$ and $|1\rangle$. Let's try to put the qubit in equal superposition(50% chance of being in state $|0\rangle$ and 50% in state |1⟩) by applying the Hadamard gate.

Similar to gates in a classical computer, there are gates in quantum computing which perform rotation operation on the qubit altering its state and probability amplitudes.

Any guesses about the position of the vector on the bloch sphere, when you apply the hadamard gate?

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1)

qc.h(0)

qc.draw(output='mpl')
```

Plot the Bloch Vector to visualize the state of the qubit after applying the Hadamard gate.

```python
state = Statevector(qc);
print(state)

plot_bloch_multivector(state)
```

$|+\rangle = \frac{|0\rangle + |1\rangle}{\sqrt2}$ and $|-\rangle = \frac{|0\rangle - |1\rangle}{\sqrt2}$ are the eigenstates of X so they are on the either sides of the X axis. Applying Hadamard gate will put the qubit in equal superposition of $|0\rangle$ and $|1\rangle$

Mathematically, all quantum gates are Unitary matrices. Here is how apply a hadamard gate will look like in a mathematical sense.

$H|0\rangle = \frac{1}{\sqrt2}
\begin{bmatrix}
1 & 1 \\
1 & -1
\end{bmatrix}
\begin{bmatrix}
1 \\
0
\end{bmatrix}=\frac{1}{\sqrt2} 
\begin{bmatrix}
1 \\
1
\end{bmatrix}=
\frac{1}{\sqrt2} \begin{bmatrix}
1 \\
0
\end{bmatrix}+\frac{1}{\sqrt2} 
\begin{bmatrix}
0 \\
1
\end{bmatrix} = \frac{|0\rangle + |1\rangle}{\sqrt2}$

<h2>Quantum Entanglement and Bell States</h2>

Quantum Entanglement is one of the bizzare phenomenon, of sub atomic particles, about which we don't yet know the why? and how? part.

Physics Perspective: We just know that if two quantum particles are entagled, then they are linked to each other even if separated by a very large distance in space.

Math Perspective: If two particles are entagled, you cannot write their state as a tensor product of two vectors(such as Bell state)

When two particles are entangled, you cannot describe about one without considering the other. Entanglement is broken when the entangled particles decohere through interaction with the environment; for example, when a measurement is made. If you measure one of the entangled particle, the state of the other is known instantaneously which seems like faster than light. But according to Einstein nothing can travel faster than time.

In 1935 Albert Einstein, Boris Podolsky, and Nathan Rosen wrote about this phenomena which is then called the famous EPR Paradox. Later Erwin Schrodinger worked on it but Einstein was not much of a supporter of this non local phenomenon.

```python
# IMPORTING REQUIRED PACKAGES
import numpy as np
from qiskit.circuit.library import IQP
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit.quantum_info import random_hermitian
from qiskit_ibm_runtime import QiskitRuntimeService

#service = QiskitRuntimeService()



# SELECTING THE LEAST BUSY QUANTUM DEVICE TO RUN THE CIRCUIT
backend = service.least_busy(operational=True, simulator=False, min_num_qubits=127)
print(backend)




# CREATING THE BELL CIRCUIT
n_qubits = 2

circuit = QuantumCircuit(2,2)
circuit.h(0)
circuit.cx(0,1)
circuit.measure_all()



# OPTIMIZING THE CIRCUIT TO THE DEVICE SELECTED
pm = generate_preset_pass_manager(optimization_level=1, backend=backend)
isa_circuit = pm.run(circuit)



# GETTING SAMPLER TO RUN THE CIRCUIT AND GET COUNTS
from qiskit_ibm_runtime import SamplerV2 as Sampler
sampler = Sampler(mode=backend)



#RUNNING THE JOB ON THE DEVICE USING SAMPLER
job = sampler.run([isa_circuit])
print(f">>> Job ID: {job.job_id()}")
print(f">>> Job Status: {job.status()}")



# GETTING RESULTS FROM THE JOB AND PLOTTING THE HISTOGRAM
result = job.result()

# Get results for the first (and only) PUB
pub_result = result[0]
print(f"Counts for the meas output register: {pub_result.data.meas.get_counts()}")

plot_histogram(pub_result.data.meas.get_counts())
```
